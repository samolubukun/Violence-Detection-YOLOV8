```python
import gradio as gr
import torch
from ultralytics import YOLO
from PIL import Image
import cv2
import numpy as np
import tempfile
import os

# Load the trained model
model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

def detect_objects(image_path):
    results = model(image_path, classes=[1])  # Assuming 'Violence' is class index 1
    for r in results:
        processed_image = r.plot()
        temp_image_path = tempfile.NamedTemporaryFile(suffix=".jpg", delete=False).name
        cv2.imwrite(temp_image_path, processed_image)
    return temp_image_path

def detect_objects_video(video_path):
    cap = cv2.VideoCapture(video_path)
    temp_video_path = tempfile.NamedTemporaryFile(suffix=".mp4", delete=False).name
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    out = cv2.VideoWriter(temp_video_path, fourcc, fps, (frame_width, frame_height))
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        results = model(frame, classes=[1])  # Filtering for 'Violence' class only
        for r in results:
            frame = r.plot()
        
        out.write(frame)
    
    cap.release()
    out.release()
    return temp_video_path

with gr.Blocks() as iface:
    gr.Markdown("## Violence Detection using YOLOv8")
    gr.Markdown("Upload an image or a video to detect violence.")
    
    with gr.Tab("Image Detection"):
        img_input = gr.Image(type="filepath")
        img_output = gr.Image(type="filepath")
        img_button = gr.Button("Detect Violence in Image")
        img_button.click(detect_objects, inputs=img_input, outputs=img_output)
    
    with gr.Tab("Video Detection"):
        video_input = gr.File(type="filepath")
        video_output = gr.Video(height=360, width=640)  # Centered, reasonable size
        video_button = gr.Button("Detect Violence in Video")
        video_button.click(detect_objects_video, inputs=video_input, outputs=video_output)

if __name__ == "__main__":
    iface.launch()
```
